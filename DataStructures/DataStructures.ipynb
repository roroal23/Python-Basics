# Types of sequences

Sequences can be classified as container or flat:
- Container sequences can hold various types of object. They store their references rather than their values. Ex: List, tuple, collections.deque
- Flat sequences can only hold one type of object, they store the actual value so they need less resources to work. Ex: str, bytes, collections.deque

# Generator expressions

Generator expressions operate similarly to list comprehension in syntax, however they save memory because they don't build a whole list, instead they generate the element one by one. You can use this to initialize a list, or to feed a loop directly.

The syntax for gen exp is the same as list comp, but with ()

```python
tup = tuple(number for number in range(0,3))
print(tup)
```

```python
for x in (el for el in range(0,5)):
    print(x*2)
```
